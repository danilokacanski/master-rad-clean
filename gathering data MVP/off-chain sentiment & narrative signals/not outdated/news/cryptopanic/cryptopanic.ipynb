```python
# CryptoPanic latest posts (cryptopanic_data sentiment)

import os, requests, pandas as pd
from pathlib import Path
from dotenv import load_dotenv

load_dotenv()
TOKEN = os.getenv("CRYPTOPANIC_TOKEN")

params = {
    "auth_token": TOKEN,
    "filter": "rising",   # 'hot','important','bullish','bearish','cryptopanic_data'
    "kind": "news",       # 'cryptopanic_data' or 'media'
    "regions": "en",
    "currencies": "BTC,ETH"
}
r = requests.get("https://cryptopanic.com/api/v1/posts/", params=params)
r.raise_for_status()
items = r.json().get("results", [])

df = pd.json_normalize(items)
out = "cryptopanic_data/"
Path(out).mkdir(parents=True, exist_ok=True)
df.to_csv(out + "cryptopanic_news.csv", index=False)

df.head(5)
```
